```python
import fitz

file_path = "/content/Haritha.resume.pdf"

doc = fitz.open(file_path)
resume_text = ""
for page in doc:
    resume_text += page.get_text()

print("--- Successfully Read Resume! ---")
print("Here is a preview of the text:")
print("-" * 30)
print(resume_text[:500])
```

```python
keywords = ["Python", "Java", "SQL", "Machine Learning", "Excel", "Project Management", "HTML", "Communication"]

found_skills = []

for skill in keywords:
    if skill.lower() in resume_text.lower():
        found_skills.append(skill)

print("--- AI Skill Extraction ---")
print(f"Skills Found: {found_skills}")
print(f"Total Skills identified: {len(found_skills)}")
```

```python
job_requirements = ["Java", "HTML", "C", "Python", "SQL", "Teamwork", "Problem-solving"]

matches = []
for req in job_requirements:
    if req.lower() in resume_text.lower():
        matches.append(req)

match_score = (len(matches) / len(job_requirements)) * 100

print(f"--- Resume Screening Report ---")
print(f"Job Requirements: {job_requirements}")
print(f"Matched Skills: {matches}")
print(f"Match Score: {match_score:.2f}%")

if match_score >= 50:
    print("Result: CANDIDATE SHORTLISTED")
else:
    print("Result: NOT A MATCH")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

job_box = widgets.Textarea(placeholder='Paste Job Description Here...', description='Job:', layout={'height': '150px', 'width': '90%'})
btn = widgets.Button(description="Run AI Screening", button_style='info')
out = widgets.Output()

def screen_it(b):
    with out:
        clear_output()
        job_text = job_box.value.lower()
        found = [skill for skill in job_requirements if skill.lower() in job_text and skill.lower() in resume_text.lower()]

        score = (len(found) / len(job_requirements)) * 100
        print(f"--- SCREENING RESULT ---")
        print(f"Match Score: {score:.1f}%")
        print(f"Skills Found: {found}")
        print("Status: PASS" if score >= 50 else "Status: FAIL")

btn.on_click(screen_it)

# 3. Show the dashboard
display(job_box, btn, out)
```
